```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Datos
datos_1 = [102, 90, 78, 40] #una matriz de correlación necesita el mismo número de filas y columnas
datos_2 = [100, 80, 70, 45]
datos_3 = [105, 94, 60, 38]
datos_4 = [107, 90, 80, 30]

#Agrupar columnas
datos = list(zip(datos_1, datos_2, datos_3, datos_4))

#Nombre de las columnas

name = ['col_1', 'col_2', 'col_3', 'col_4']
```

```python
#Dataframe

df = pd.DataFrame(data = datos, columns = name)
df
```

```python
ccor = df.corr('pearson')
ccor #Matriz jabiana. La diagonal ppal en 1
```

```python
plt.figure(figsize = (10, 10))
sns.heatmap(ccor, annot = True)
plt.show()
```
